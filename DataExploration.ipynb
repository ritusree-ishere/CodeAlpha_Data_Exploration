```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load iris.csv into pandas DataFrame
iris = pd.read_csv("iris.csv")
```

```python
#Data-points and features of the data table 
print (iris.shape)
```

```python
#the column names of the data table displayed
print (iris.columns)
```

```python
iris['Species'].value_counts()
#It's a balanced dataset
```

```python
#2-D scatter plot
iris.plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm')
plt.show();
```

```python
sns.set_style("whitegrid") ;

g = sns.FacetGrid(iris, hue="Species", height=4)
g.map(plt.scatter, "SepalLengthCm", "SepalWidthCm")
g.add_legend() ;

plt.show() ;

#Blue points can be separated easily.
#From green and orange by drawing a line.
#But green and orange lines cannot be seaprated easily
```
